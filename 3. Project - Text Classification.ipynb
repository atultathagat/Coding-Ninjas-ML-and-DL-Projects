## Extracting text and its corresponding categories from dataset

```python
def extract_text_and_category():
    texts=[] #list containing texts of different articles 
    categories=[] # list containing categories to which these articles belong
    cnt=1
    for category in os.listdir(src_path):
        file_path=os.path.join(src_path,category)
        for file_name in os.listdir(file_path):
            path=os.path.join(file_path,file_name)
            file_obj=open(path,'r')
            text=file_obj.read().strip() #reading text from different files
            texts.append(text)
            categories.append(category)
            print('Total documents extracted:',cnt)
            cnt+=1
    return texts,categories        
```

```python
texts,categories=extract_text_and_category()
```

## Getting all the words in different articles after removing stopwords, non alphabet words and the words having length less than or equal to 3

```python
def get_all_words(texts):
    stop_words=get_stop_words('english')
    all_words=[] #list containing different words
    for cnt,text in enumerate(texts):
        words=re.findall(r'[^\s!,.?":;0-9]+', text) #splitting sentence into words
        for word in words:
            word=word.lower()
            if word.isalpha() and (not word in stop_words) and len(word)>3:
                all_words.append(word)
    return all_words            
```

```python
all_words=get_all_words(texts)
```

## Calculating frequency of different words and finding top 2000 words having maximum frequency

```python
#function returns words with its frequency
def get_frequency_of_the_words(all_words):
    words,frequency=np.unique(all_words,return_counts=True)
    #sorting the frequency of words in decreasing order
    indices=np.argsort(frequency)[::-1]
    words=words[indices]
    frequency=frequency[indices]
    return words,frequency
```

```python
words,frequency=get_frequency_of_the_words(all_words)
```

```python
#plotting the graph of words vs its corresponding frequency
plt.plot(frequency)
plt.axis([0,2500,0,12000])
plt.xlabel('Words-->')
plt.ylabel('Frequency of words-->')
plt.title('Words vs Frequency of words')
plt.show()
```

```python
#extracting top 2000 words having maximum frequency
def extract_top_2000_words(words,frequency):
    words=words[:2000]
    frequency=frequency[:2000]
    print('Top 2000 words in the documents along with their frequency:\n')
    for i in range(2000):
        print(words[i],':',frequency[i])
    return words,frequency    
```

```python
#printing top 2000 words along with its frequency
words,frequency=extract_top_2000_words(words,frequency)
```

```python
#finding a dictionary which stores word as key and an unique label (number) to each word as value
def label_words(words):
    dict_word_to_idx={}
    for idx,w in enumerate(words):
        dict_word_to_idx[w]=idx
    return dict_word_to_idx    
```

```python
dict_word_to_idx=label_words(words)
```

```python
#getting the data in the required format on which we can train our model
def get_data(texts,M,N): #M is number of data points and N is number of features in the data point
    x=np.zeros((M,N)) # a 2D matrix having M rows and N columns, where each row represents a data point having frequency of different words into it
    for i,text in enumerate(texts):
        w=re.findall(r'[^\s!,.?":;0-9]+',text)
        for word in w:
            if word in words:
                x[i][dict_word_to_idx[word]]+=1
        print("Total documents processed:",i+1)
    return x     
```

```python
M=len(texts)
N=len(words)
x=get_data(texts,M,N)
x.shape
```

## Implementing Naive Bayes from Scratch

```python
#finding trained dictionary based on the values of x train and y train
def fit(x_train,y_train):
    unique_categories=set(categories)
    trained_dict={}
    N=len(x_train[0])
    trained_dict['sum_total']=len(y_train)
    for category in unique_categories:
        indices=np.where(y_train==category)
        y_train_category=y_train[indices]
        trained_dict[category]={}
        trained_dict[category]['total']=len(y_train_category)
        x_train_category=x_train[indices]
        total_words=0
        for i in range(N):
                total=x_train_category[:,i].sum()      
                trained_dict[category][i]=total
                total_words+=total
        trained_dict[category]['total_words']=total_words                 
    return trained_dict
```

```python
#finding the probability of a training data point x belonging to a cateogry y
def probability_x_belongs_to_y(x,y,trained_dict):
    class_probability=np.log(trained_dict[y]['total'])-np.log(trained_dict['sum_total'])
    N=len(x)
    for i in range(N):
        if x[i]>0:
            probability=np.log(trained_dict[y][i]+1)-np.log(trained_dict[y]['total_words']+N)
            class_probability+=probability
    return class_probability        
```

```python
#predicting the class (category) for a single data point x
def predict_single_data_point(x,trained_dict):
    y_pred=''
    max_probability=-1
    unique_categories=set(categories)
    for y in unique_categories:
        probability=probability_x_belongs_to_y(x,y,trained_dict)
        if max_probability==-1:
            y_pred=y
            max_probability=probability
        elif probability>max_probability:
            y_pred=y
            max_probability=probability
    return y_pred        
```

```python
#predicting the classes (categories) for all the data points
def predict_all_data_points(x_test,trained_dict):
    y_pred_scratch=[]
    for x in x_test:
        y_pred=predict_single_data_point(x,trained_dict)
        y_pred_scratch.append(y_pred)
    return np.array(y_pred_scratch)    
```

```python
trained_dict=fit(x_train,y_train) #getting the trained dictionary
y_pred_scratch=predict_all_data_points(x_test,trained_dict)
```

```python
#confusion matrix for implementation of naive bayes from scratch
confusion_matrix(y_test,y_pred_scratch)
```

```python
#classification report for implementation of naive bayes from scratch
print(classification_report(y_test,y_pred_scratch))
```

## Comparison of result of Sklearn Implementation of Naive Bayes vs Implementation from scratch

```python
print('Accuracy with sklearn implementation:',accuracy_score(y_test,y_pred_sklearn))
print('Accuracy with implementation from scratch:',accuracy_score(y_test,y_pred_scratch))
```
